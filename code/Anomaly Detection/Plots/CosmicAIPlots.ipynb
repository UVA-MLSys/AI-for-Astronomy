```python
from matplotlib.ticker import LogFormatterSciNotation 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load data
data = pd.read_csv("Cloud for CosmicAI(CURRENT - vary batch size).csv")
data = data[(data['average throughput']>0)&(data['Processor']!="Dell XPS Server")]

# Custom color palette for the processors
custom_palette = {
    "Personal Laptop": (0.1, 0.2, 0.6),   # Muted blue in RGB (normalized to 0-1 range)
    # "Dell XPS Server": (0.9, 0.5, 0.2),   # Muted orange in RGB
    "Rivanna HPC GPU": (0.8, 0.6, 0.1),   # Muted gold in RGB 
    "Rivanna HPC CPU": (0.2, 0.6, 0.2),   # Muted green in RGB
    "CAI": (0.5, 0.2, 0.6)               # Muted purple in RGB
}

# Create the plot
plt.figure(figsize=(9, 6))

# Loop over each processor to adjust marker size, line width, and color
for processor in data['Processor'].unique():
    processor_data = data[data['Processor'] == processor]
    
    # Set marker size and line width based on processor
    if processor in ["Rivanna HPC GPU", "CAI"]:
        marker_size = 14
        line_width = 3.75
        marker_style = "o"  # Circle marker for these processors
    else:
        marker_size = 10
        line_width = 2.5
        marker_style = "o"  # Square marker for others

    # Plot the line with custom color and properties
    ax = sns.lineplot(
        data=processor_data, 
        x='batch size', 
        y='average throughput', 
        markers=True, 
        dashes=False, 
        markersize=marker_size,  # Adjust marker size
        linewidth=line_width,    # Adjust line width
        color=custom_palette[processor],  # Use custom color
        marker=marker_style,     # Explicitly set marker style
        label=processor.replace('Rivanna ', ''),         # Ensure proper label for annotation
        legend=False             # Disable automatic legend creation
    )
    
    # Set log scale for y-axis
    ax.set_yscale("log")

# Adjust the font size for axis ticks
plt.tick_params(axis='both', which='major', labelsize=14)
plt.gca().yaxis.get_offset_text().set_fontsize(14)

# Add labels and title
plt.xlabel('Batch Size', fontsize=16)
plt.ylabel('Throughput (bps) [log-scaled]', fontsize=16)
plt.title('Throughput (bps) vs Batch Size by Processor', fontsize=20, fontweight='bold')

# Force labels to align at Batch Size = 400
target_x = 400
handles, labels = plt.gca().get_legend_handles_labels()
for line, label in zip(plt.gca().get_lines(), labels):
    x_data = line.get_xdata()
    y_data = line.get_ydata()
    
    # Interpolate y-value at target_x
    if target_x in x_data:
        y_at_target = y_data[np.where(x_data == target_x)[0][0]]
    else:
        y_at_target = np.interp(target_x, x_data, y_data)
                  
# Reverse the legend order if necessary
plt.legend(
    handles=handles[::-1], labels=labels[::-1], 
    title='Processor', fontsize=12, title_fontsize=14, 
    bbox_to_anchor=(0.48, 0.85),
    ncols=2,
    alignment='center')

# Add horizontal grid lines only
plt.grid(axis='y', color='gray', linewidth=0.5, alpha=0.5)

# Save the figure
plt.savefig("batchsize_throughput_log_scaled.jpg", dpi=200)
```
