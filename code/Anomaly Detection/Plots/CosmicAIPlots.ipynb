```python
import seaborn as sns

# Load the data
data = pd.read_csv("FinalDataVaryResults.csv")

# Define a custom RGB color palette for each processor (using muted colors)
custom_palette = {
    "Personal Laptop": (0.1, 0.2, 0.6),   # Muted blue in RGB (normalized to 0-1 range)
    "Dell XPS Server": (0.9, 0.5, 0.2),   # Muted orange in RGB
    "Rivanna HPC GPU": (0.8, 0.6, 0.1),   # Muted gold in RGB 
    "Rivanna HPC CPU": (0.2, 0.6, 0.2),   # Muted green in RGB
    "CAI": (0.5, 0.2, 0.6)               # Muted purple in RGB
}

# Set up the plot
plt.figure(figsize=(10, 6))

# Define a function to handle different marker sizes and line widths
def plot_line_with_marker_size_and_linewidth(data, x, y, hue, style, custom_palette):
    for processor in data[hue].unique():
        processor_data = data[data[hue] == processor]
        
        # Set marker size and line width based on the processor
        marker_size = 14 if processor in ["Rivanna HPC GPU", "CAI"] else 10
        line_width = 3.75 if processor in ["Rivanna HPC GPU", "CAI"] else 2.5
        
        sns.lineplot(
            data=processor_data,
            x=x,
            y=y,
            hue=hue,
            style=style,
            markers=True,
            dashes=False,
            markersize=marker_size,  # Set marker size based on processor
            linewidth=line_width,    # Set line width based on processor
            palette=custom_palette,
            label=processor,
            legend=False  # Disable the automatic legend creation for this line
        )

# Plot each line with the appropriate marker size and line width
plot_line_with_marker_size_and_linewidth(data, 'Dataset Size (GB)', 'Average Time (s)', 'Processor', 'Processor', custom_palette)

# Add labels and title
plt.tick_params(axis='both', which='major', labelsize=14)
plt.xlabel('Dataset Size (GB)', fontsize=16)
plt.ylabel('Total Inference Time (s)', fontsize=16)
plt.title('Total Inference Time (s) vs Dataset Size (GB) by Processor', fontsize=20, fontweight='bold')

# Create a custom legend with unique entries
handles, labels = plt.gca().get_legend_handles_labels()

# Reverse the legend order if necessary
plt.legend(handles=handles[::-1], labels=labels[::-1], title='Processor', fontsize=11, title_fontsize=12)

# Add horizontal grid lines only
plt.grid(axis='y', color='gray', linewidth=0.5, alpha=0.5)

plt.savefig("datasize_time.png")
```

```python
# Create a custom purple color palette with subtly varying shades of purple
import numpy as np
import seaborn as sns
import colorsys
from matplotlib.ticker import FuncFormatter

# Function to adjust lightness slightly
def adjust_lightness(color, amount=1):
    # Convert RGB to HLS using colorsys
    hls_color = colorsys.rgb_to_hls(*color)
    
    # Adjust the lightness (keep hue and saturation fixed)
    lightness = np.clip(hls_color[1] + amount, 0, 1)
    
    # Convert back to RGB
    return colorsys.hls_to_rgb(hls_color[0], lightness, hls_color[2])

# Generate slightly different shades of purple
unique_partitions = df['partition (MB)'].unique()
num_partitions = len(unique_partitions)

# Generate purple palette with slight variations in lightness
purple_palette = [adjust_lightness(base_purple, amount=0.15 * i) for i in range(num_partitions)]
partition_color_map = dict(zip(unique_partitions, purple_palette))

# Set up the plot
plt.figure(figsize=(10, 6))

# Plot each partition with its respective color
for index, (partition, color) in enumerate(partition_color_map.items()):
    partition_data = df[df['partition (MB)'] == partition]
    sns.lineplot(
        data=partition_data,
        x='data (GB)',
        y='inference_duration (s)',
        label=f'Partition {partition} MB',
        marker=markers[index % len(markers)],
        linewidth=2.75,
        markersize=12,
        color=color,
        errorbar=None,  # Disable confidence interval shading
    )

# Set log scale for y-axis
#plt.yscale('log')

# Custom formatter for y-axis labels to ensure consistent font size
def y_axis_formatter(value, tick_number):
    return f"{value:.1e}"

#plt.gca().yaxis.set_major_formatter(FuncFormatter(y_axis_formatter))
plt.tick_params(axis='both', which='major', labelsize=14)  # Increase tick label font size

plt.xlabel('Dataset Size (GB)', fontsize=16)
plt.ylabel('Inference Time (s)', fontsize=16)
plt.title(
    "CAI Varying Partition Sizes\nInference Time (s) vs. Data Size (GB)",
    fontsize=20,
    fontweight='bold'
)

# Annotate labels above each line
for line, label in zip(plt.gca().get_lines(), partition_color_map.keys()):
    x_data = line.get_xdata()
    y_data = line.get_ydata()

    # Get the maximum y value and its corresponding x value for annotation
    max_y = max(y_data)
    max_x = x_data[np.argmax(y_data)]
    
# Reverse the order of the legend and move it to the bottom right
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(
    handles=handles[::-1],  # Reverse the order of handles
    labels=labels[::-1],    # Reverse the order of labels
    fontsize=12,
    title='Partition Sizes (MB)',
    title_fontsize=14,
    # bbox_to_anchor=(0.45, 0.1),
    loc='lower right',  # Position the legend at the bottom right corner,
    ncols=2
)

# Add horizontal grid lines only
plt.grid(axis='y', color='gray', linewidth=0.5, alpha=0.5)

# Save the figure
plt.savefig("cai_inference_time.jpg", dpi=200)
```
